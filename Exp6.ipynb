```python
date_str =(input("Enter a date in yyyy-mm-dd :"))



parts = date_str.split('-')

if len(parts) == 3 and all(p.isdigit() for p in parts):
    year = int(parts[0])
    month = int(parts[1])
    day = int(parts[2])

    print(f"Original: {year}-{month:02d}-{day:02d}")


    day += 1

    # Month days array
    days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if day > days_in_month[month-1]:
        day = 1
        month += 1
        if month > 12:
            month = 1
            year += 1

    print(f"Next day: {year}-{month:02d}-{day:02d}")
else:
    print("Invalid format! Use YYYY-MM-DD (e.g., 2026-02-09)")
```
